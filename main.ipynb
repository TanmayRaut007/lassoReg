```python
import pandas as pd
data = pd.read_csv("houseData.csv")
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
data = data[cols_to_use]
```

```python
data.head(10)
```

```python
data.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
data[cols_to_fill_zero] = data[cols_to_fill_zero].fillna(0)
```

```python
data['Landsize'] = data['Landsize'].fillna(data.Landsize.mean())
data['BuildingArea'] = data['BuildingArea'].fillna(data.BuildingArea.mean())
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

```python

X = data.drop('Price', axis=1)
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train ,y_test = train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=50, max_iter=10000, tol=0.1)
lasso_reg.fit(X_train, y_train)  # âœ… Fit on training data
score = lasso_reg.score(X_test, y_test)
print("Lasso RÂ² score on test set:", score)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=50, max_iter=10000, tol=0.1)
ridge_reg.fit(X_train, y_train)  # âœ… Train on training data
print("Ridge RÂ² score on test set:", ridge_reg.score(X_test, y_test))
```
